```python
flights = pd.read_csv("flights.csv")
```

```python
flights.head()
```

```python
flights["celsius"] = (flights.temp - 32) * 5/9
```

```python
flights.head()
```

```python
jfk = flights.loc[flights['origin'] == 'JFK']
```

```python
jfk
```

```python
jfk_grouped = jfk.groupby(['month', "day"]).mean()
print(jfk_grouped)
```

```python
jfk_grouped['match'] = jfk_grouped.celsius.gt(jfk_grouped.celsius.shift())
jfk_grouped.head(10)
```

```python
jfk_grouped.sort_values('celsius', ascending=False).head(5)
```

```python
A = pd.read_table(url1, sep=',')
A.head()
B = pd.read_table(url2, sep=',')
B.head()
C = pd.read_table(url3, sep=',')
C.head()
```

```python
D = pd.merge(A, B, on='Name', how='outer')
print(D)
```

```python
E = pd.merge(D, C, on="Name", how="outer")
print(E)
```

```python
F = pd.merge(A, B, on='Name', how='inner')
print(F)
```

```python
G = pd.merge(A, C, on='Name', how='inner')
print(G)
```

```python
H = pd.merge(A, B, on='Name', how='left')
print(H)
```
