```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/DavidePosillipo/ProbabilityAndStatistics/master/Assignment_2/wines_properties.csv"
```

```python
wine = pd.read_table(url, sep=',')
```

```python
wine.head()
```

```python
wine_id = wine.columns[-1]
wine_var = wine.columns[0:13]
wine_id
wine_var
```

```python
pc_id = wine.loc[:, wine_id].values
pc_var = wine.loc[:, wine_var].values
print(pc_id)
print(pc_var)
```

```python
from sklearn.preprocessing import StandardScaler
pc_var_std = StandardScaler().fit_transform(pc_var)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

PCs = pca.fit_transform(pc_var_std)
print(PCs)
```

```python
PCs_an = pca.components_
print(PCs_an)
```

```python
# Use quiver to generate the basic plot
fig = plt.figure(figsize=(12,12))
plt.quiver(np.zeros(PCs_an.shape[1]), np.zeros(PCs_an.shape[1]),
           PCs_an[0,:], PCs_an[1,:], 
           angles='xy', scale_units='xy', scale=1)

# Add labels based on feature names (here just numbers)
feature_names = np.arange(PCs_an.shape[1])
for i,j,z, in zip(PCs_an[1,:]+0.02, PCs_an[0,:]+0.02, feature_names):
    plt.text(j, i, z, ha='center', va='center')

# Add unit circle
circle = plt.Circle((0,0), 1, facecolor='none', edgecolor='b')
plt.gca().add_artist(circle)

# Ensure correct aspect ratio and axis limits
plt.axis('equal')
plt.xlim([-1.0,1.0])
plt.ylim([-1.0,1.0])

# Label axes
plt.xlabel('PC 1')
plt.ylabel('PC 2')

# Done
plt.show()
print(fig)
```

## PC1 mainly includes variables 5, 6, 8 (positively) : Total_Phenols, Flavanoids, and Proanthocyanins. And 3 and 7 to a lesser extent (negatively): Ash_Alcanity, Nonflavanoid_Phenols. 
## While PC2 includes var.  0, 9, 12 (negatively): Alcohol, Color_intensity, and Proline

```python
pc_wine = pd.DataFrame(PCs, columns=['PC1', 'PC2'])
print(pc_wine)
```

## Hierarchical cluster

```python
from scipy.spatial.distance import pdist, squareform
```

```python
row_dist = pd.DataFrame(squareform(pdist(pc_var_std, metric='euclidean')),)
row_dist
```

```python
from scipy.cluster.hierarchy import linkage

row_clusters = linkage(row_dist, method='complete', metric='euclidean')
pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=['cluster %d' % (i + 1)
                    for i in range(row_clusters.shape[0])])
```

```python
from scipy.cluster.hierarchy import dendrogram

row_dendr = dendrogram(row_clusters)
plt.tight_layout()
plt.ylabel('Euclidean distance')

plt.show()
```

## I would guess 4 or 5 clusters

## KMEANS



```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4, 
            init='random', 
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)

y_km = km.fit_predict(pc_var_std)
print(y_km)
```

```python
# cluster 1
plt.scatter(pc_var_std[y_km == 0, 0],
            pc_var_std[y_km == 0, 1],
            s=50, c='lightgreen',
            marker='s', edgecolor='black',
            label='cluster 1')
# cluster 2
plt.scatter(pc_var_std[y_km == 1, 0],
            pc_var_std[y_km == 1, 1],
            s=50, c='orange',
            marker='o', edgecolor='black',
            label='cluster 2')
# cluster 3
plt.scatter(pc_var_std[y_km == 2, 0],
            pc_var_std[y_km == 2, 1],
            s=50, c='lightblue',
            marker='v', edgecolor='black',
            label='cluster 3')
# cluster 4
plt.scatter(pc_var_std[y_km == 3, 0],
            pc_var_std[y_km == 3, 1],
            s=50, c='yellow',
            marker='v', edgecolor='black',
            label='cluster 4')

# centroids 
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250, marker='*',
            c='red', edgecolor='black',
            label='centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
from matplotlib import cm
from sklearn.metrics import silhouette_samples
from sklearn import metrics

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]

silhouette_vals = silhouette_samples(pc_var_std, y_km, metric='euclidean')

y_ax_lower, y_ax_upper = 0, 0
yticks = []

for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, 
             edgecolor='none', color=color)

    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--") 

plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

plt.tight_layout()
plt.show()
```
